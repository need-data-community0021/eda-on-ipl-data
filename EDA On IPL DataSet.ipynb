```python
#loading data sets
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
#loading data set
ipl=pd.read_csv("D:\\Drivers\\matches.csv")
ipl
```

```python
#Checking the head of the dataset
ipl.head()
```

```python
#check dataset information 
ipl.info()
```

```python
#Drop the missing value in the dataset using "dropna" 
ipl.dropna()
```

```python
#looking the shape of the data sets
ipl.shape
```

```python
#looking of the most man of the match award
ipl['player_of_match'].value_counts()
```

```python
#looking the top 10 players with most of thr match award
ipl['player_of_match'].value_counts()[0:10]
```

```python
#cheak the top 5 player man of the match 
list(ipl['player_of_match'].value_counts()[0:5].keys())
```

```python
#Making the bar plot 
plt.figure(figsize=(8,5))
plt.bar(list(ipl['player_of_match'].value_counts()[0:5].keys()),list(ipl['player_of_match'].value_counts()[0:5]),color="g")
plt.show()
```

```python
#getting the frequency of result column 
ipl['result'].value_counts()
```

```python
#Finding out the number of toss wins W.R.T team
ipl['toss_winner'].value_counts()
```

```python
#Extracting the record where a team won  batting
batting_first=ipl[ipl['win_by_runs']!=0]
```

```python
batting_first.head()
```

```python
#Amking a histogram 
plt.figure(figsize=(5,4))
plt.hist(batting_first['win_by_runs'])
plt.title("Distribution of Runs")
plt.xlabel("Runs")
plt.show()
```

```python
#Finding out the number of wins w.r.t each team after team after batting first 
batting_first['winner'].value_counts()
```

```python
#making a bar-plot for top 3 teams with most wins after batting first 
plt.figure(figsize=(6,6))
plt.bar(list(batting_first['winner'].value_counts()[0:3].keys()),list(batting_first['winner'].value_counts()[0:3]),color=["blue","grey"])
```

```python
#Making a pie chart
plt.figure(figsize=(7,7))
plt.pie(list(batting_first['winner'].value_counts()),labels=list(batting_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
#Extracting those record where a team has won after batting second
batting_second=ipl[ipl['win_by_wickets']!=0]
```

```python
batting_second.head()
```

```python
#making the histogram for frequency of wins w.r.t of wicket
plt.figure(figsize=(8,5))
plt.hist(batting_second['win_by_wickets'],bins=30)
plt.show()
```

```python
#Finding out the number of wins w.r.t each time after batting second 
batting_second['winner'].value_counts()
```

```python
#making a bar plot for top-3 teams with most wins after batting second
plt.figure(figsize=(7,7))
plt.bar(list(batting_second['winner'].value_counts()[0:3].keys()),list(batting_second['winner'].value_counts()[0:3]),color=["purple","blue","red"])
plt.show()
```

```python
#making a pie chart for distribution of most wins after batting second
plt.figure(figsize=(7,7))
plt.pie(list(batting_second['winner'].value_counts()),labels=list(batting_second['winner'].value_counts().keys()),autopct="%0.1f%%")
plt.show()
```
